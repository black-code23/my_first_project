```python
label_encoder = LabelEncoder()
india_crop_merged1['crop'] = label_encoder.fit_transform(india_crop_merged1['crop'])
india_crop_merged1['District'] = label_encoder.fit_transform(india_crop_merged1['District'])
india_crop_merged1['typesofsoil'] = label_encoder.fit_transform(india_crop_merged1['typesofsoil'])
imputer = KNNImputer(n_neighbors = 5)
imputer.fit(india_crop_merged1)
india_crop_merged1_imputed = imputer.transform(india_crop_merged1)
scaler = StandardScaler()
india_crop_merged1_scaled = scaler.fit_transform(india_crop_merged1_imputed)
y = india_crop_merged1_scaled[:,-1]
x = india_crop_merged1_scaled[:,:11]


```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 236)
```

```python
model = RandomForestRegressor()
model_fit = model.fit(x_train,y_train)
y_pred = model_fit.predict(x_test)
score = r2_score(y_test,y_pred)
score1 = mean_squared_error(y_test,y_pred)
score2 = model.score(x_test,y_test)
print(score)
print(score1)
print(score2)
```
